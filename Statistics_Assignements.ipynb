10.Explain nominal, ordinal, interval, and ratio levels of measurement?
->Nominal: Categories with no order (e.g., gender, blood type).

Ordinal: Categories with a meaningful order but unequal gaps (e.g., rankings, satisfaction levels).

Interval: Numeric data with equal intervals but no true zero (e.g., temperature in Celsius).

Ratio: Numeric data with equal intervals and a true zero, allowing all math operations (e.g., height, weight, age).

16.Define and explain the term range in statistics?

18.What is skewness in a dataset?

19.What does it mean if a dataset is positively or negatively skewed?
Skewness measures the asymmetry of data distribution.  
- **Positive skew:** Right tail longer, data concentrated on left.  
- **Negative skew:** Left tail longer, data concentrated on right.

20. Define and explain kurtosis?
Kurtosis measures the "tailedness" of data distribution.  
- High kurtosis: heavy tails, more outliers.  
- Low kurtosis: light tails, fewer outliers.

21. What is the purpose of covariance?
->Measures how two variables vary together:  
- Positive covariance: variables increase/decrease together.  
- Negative covariance: one increases, the other decreases.


22.What does correlation measure in statistics?

- Range: -1 to +1 (standardized covariance).

23.What is the difference between covariance and correlation?
- Covariance: indicates direction, depends on units, unstandardized.  
- Correlation: standardized covariance, shows strength & direction, unitless.

```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Calculate mean, median, and mode
data = [12, 15, 12, 18, 19, 15, 12, 20]
mean_val = np.mean(data)
median_val = np.median(data)
mode_result = stats.mode(data, keepdims=True)
mode_val = mode_result.mode[0]
print(f"Mean: {mean_val}, Median: {median_val}, Mode: {mode_val}")

# 2. Variance and standard deviation
variance = np.var(data, ddof=1)  # sample variance
std_dev = np.std(data, ddof=1)   # sample std deviation
print(f"Variance: {variance}, Standard Deviation: {std_dev}")

# 3. Classify dataset into nominal, ordinal, interval, ratio
# Nominal: categories without order
nominal = ['red', 'blue', 'green', 'blue']
# Ordinal: ordered categories
ordinal = ['low', 'medium', 'high', 'medium']
# Interval: numeric scales, no true zero (e.g., temperature in Celsius)
interval = [10, 20, 30, 25]
# Ratio: numeric scales with true zero (e.g., weight)
ratio = [0, 50, 100, 150]
print(f"Nominal: {nominal}\nOrdinal: {ordinal}\nInterval: {interval}\nRatio: {ratio}")

# 4. Random sampling and stratified sampling
df = pd.DataFrame({'Category': ['A', 'A', 'B', 'B', 'C', 'C', 'C', 'A'],
                   'Value': [10, 15, 10, 20, 10, 20, 30, 15]})

# Random sampling
random_sample = df.sample(n=4, random_state=1)
print("Random sample:\n", random_sample)

# Stratified sampling: sample from each category proportional to their frequency
stratified_sample = df.groupby('Category', group_keys=False).apply(lambda x: x.sample(frac=0.5, random_state=1))
print("Stratified sample:\n", stratified_sample)

# 5. Function to calculate range
def data_range(d):
    return max(d) - min(d)

print(f"Range of data: {data_range(data)}")

# 6. Histogram to visualize skewness
skewed_data = np.random.chisquare(df=2, size=1000)
plt.hist(skewed_data, bins=30, edgecolor='black')
plt.title("Histogram showing positive skewness")
plt.show()

# 7. Calculate skewness and kurtosis
skewness = stats.skew(skewed_data)
kurtosis = stats.kurtosis(skewed_data)
print(f"Skewness: {skewness}, Kurtosis: {kurtosis}")

# 8. Generate positive and negative skewness
pos_skew = np.random.exponential(scale=2, size=1000)
neg_skew = -np.random.exponential(scale=2, size=1000)

plt.subplot(1,2,1)
plt.hist(pos_skew, bins=30, color='blue')
plt.title("Positive Skewness")

plt.subplot(1,2,2)
plt.hist(neg_skew, bins=30, color='red')
plt.title("Negative Skewness")

plt.tight_layout()
plt.show()

# 9. Covariance between two datasets
x = [1, 2, 3, 4, 5]
y = [2, 4, 5, 4, 5]
cov_matrix = np.cov(x, y)
cov_xy = cov_matrix[0,1]
print(f"Covariance between x and y: {cov_xy}")

# 10. Correlation coefficient
corr_coef = np.corrcoef(x, y)[0,1]
print(f"Correlation coefficient between x and y: {corr_coef}")

# 11. Scatter plot for relationship visualization
plt.scatter(x, y)
plt.title("Scatter Plot of x vs y")
plt.xlabel("x")
plt.ylabel("y")
plt.show()

# 12. Simple random sampling vs systematic sampling
population = list(range(1, 21))

# Simple random sampling (sample 5)
srs = np.random.choice(population, size=5, replace=False)
print("Simple random sample:", srs)

# Systematic sampling (every 4th item)
k = 4
start = np.random.randint(0, k)
systematic_sample = population[start::k]
print("Systematic sample:", systematic_sample)

# 13. Mean, median, mode of grouped data
grouped_data = pd.DataFrame({
    'class_interval': ['0-10', '10-20', '20-30', '30-40'],
    'frequency': [5, 8, 12, 5]
})

# Calculate midpoints
grouped_data['midpoint'] = grouped_data['class_interval'].apply(lambda x: (int(x.split('-')[0]) + int(x.split('-')[1])) / 2)

# Mean
mean_grouped = np.sum(grouped_data['midpoint'] * grouped_data['frequency']) / grouped_data['frequency'].sum()
print(f"Grouped data mean: {mean_grouped}")

# Median and mode would typically require cumulative frequency and modal class calculations, which can be complex for this short snippet

# 14. Simulate data and calculate central tendency and dispersion
sim_data = np.random.normal(loc=50, scale=10, size=1000)
print(f"Simulated data mean: {np.mean(sim_data)}")
print(f"Simulated data median: {np.median(sim_data)}")
print(f"Simulated data std deviation: {np.std(sim_data, ddof=1)}")


# Generate sample dataset
np.random.seed(0)
data = pd.DataFrame({
    'A': np.random.normal(50, 10, 100),
    'B': np.random.normal(60, 20, 100),
    'C': np.random.gamma(shape=2, scale=2, size=100)
})

# 15. Summary descriptive statistics using pandas
print("Descriptive statistics:\n", data.describe())

# 16. Boxplot to understand spread and outliers
plt.figure(figsize=(8,5))
sns.boxplot(data=data)
plt.title('Boxplot for dataset columns')
plt.show()

# 17. Calculate Interquartile Range (IQR) for each column
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1
print("\nInterquartile Range (IQR):\n", IQR)

# 18. Z-score normalization (standardization)
z_scores = stats.zscore(data)
print("\nFirst 5 rows of Z-score normalized data:\n", pd.DataFrame(z_scores, columns=data.columns).head())

# Explanation: Z-score normalization rescales data to have mean=0 and std=1, useful for comparing datasets with different units.

# 19. Compare two datasets using their standard deviations
std_dev = data.std()
print("\nStandard deviations:\n", std_dev)

# 20. Visualize covariance using heatmap
cov_matrix = data.cov()
plt.figure(figsize=(6,5))
sns.heatmap(cov_matrix, annot=True, cmap='coolwarm')
plt.title('Covariance Matrix Heatmap')
plt.show()

# 21. Correlation matrix heatmap using seaborn
corr_matrix = data.corr()
plt.figure(figsize=(6,5))
sns.heatmap(corr_matrix, annot=True, cmap='BrBG')
plt.title('Correlation Matrix Heatmap')
plt.show()

# 22. Variance and standard deviation computations
variance = data.var()
std_dev = data.std()
print("\nVariance:\n", variance)
print("\nStandard Deviation:\n", std_dev)

# 23. Visualize skewness and kurtosis
skewness = data.skew()
kurtosis = data.kurtosis()
print("\nSkewness:\n", skewness)
print("\nKurtosis:\n", kurtosis)

plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
sns.barplot(x=skewness.index, y=skewness.values)
plt.title('Skewness per Column')

plt.subplot(1,2,2)
sns.barplot(x=kurtosis.index, y=kurtosis.values)
plt.title('Kurtosis per Column')
plt.show()

# 24. Pearson and Spearman correlation coefficients
pearson_corr = data.corr(method='pearson')
spearman_corr = data.corr(method='spearman')
print("\nPearson Correlation:\n", pearson_corr)
print("\nSpearman Correlation:\n", spearman_corr)
```
